```python
import pickle as pck
import warnings
warnings.filterwarnings("ignore")
```

```python
load_model = pck.load(open("knn_model.sav", 'rb'))
input_array = load_model.n_features_in_       #feature_names_in_ variable is used to know the expected parameters in the model
print(input_array)
```

```python
input_list = []
for i in range(input_array):
    prediction_input = int(input(f"Enter Input-{i+1}:"))
    input_list.append(prediction_input)
```

```python
print("Future Prediction =", load_model.predict([input_list]))
```
